```python
import mysql.connector
import tkinter as tk
from tkinter import messagebox

# Function to fetch and display databases
def show_databases():
    try:
        # Connect to MySQL
        myconn = mysql.connector.connect(host="localhost", user="root", passwd="12345")
        cur = myconn.cursor()

        # Execute query to show databases
        cur.execute("SHOW DATABASES")

        # Clear previous entries
        databases_listbox.delete(0, tk.END)

        # Fetch and display databases
        for db in cur:
            databases_listbox.insert(tk.END, db[0])

        # Close connection
        myconn.close()

    except mysql.connector.Error as err:
        messagebox.showerror("Error", f"Database connection failed!\n{err}")

# Function to fetch and display tables for the selected database
def show_tables():
    try:
        # Get the selected database
        selected_db = databases_listbox.get(tk.ACTIVE)
        
        if not selected_db:
            messagebox.showwarning("No Database Selected", "Please select a database from the list.")
            return
        
        # Connect to the selected database
        myconn = mysql.connector.connect(host="localhost", user="root", passwd="12345", database=selected_db)
        cur = myconn.cursor()

        # Execute query to show tables
        cur.execute("SHOW TABLES")

        # Clear previous entries
        tables_listbox.delete(0, tk.END)

        # Fetch and display tables
        for table in cur:
            tables_listbox.insert(tk.END, table[0])

        # Close connection
        myconn.close()

    except mysql.connector.Error as err:
        messagebox.showerror("Error", f"Failed to fetch tables: {err}")

# Creating GUI window
root = tk.Tk()
root.title("MySQL Viewer")
root.geometry("500x400")

# Button to fetch databases
fetch_button = tk.Button(root, text="Show Databases", command=show_databases, font=("Arial", 12), bg="lightblue")
fetch_button.pack(pady=10)

# Listbox to display databases
databases_listbox = tk.Listbox(root, width=50, height=10, font=("Arial", 10))
databases_listbox.pack(pady=10)

# Button to fetch tables
show_tables_button = tk.Button(root, text="Show Tables", command=show_tables, font=("Arial", 12), bg="lightgreen")
show_tables_button.pack(pady=10)

# Listbox to display tables
tables_listbox = tk.Listbox(root, width=50, height=10, font=("Arial", 10))
tables_listbox.pack(pady=10)

# Run the GUI loop
root.mainloop()
```
